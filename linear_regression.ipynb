```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.random.normal(0,1,100)
Y=np.random.uniform(0,1,100)

X.shape, Y.shape
```

```python
plt.scatter(X,Y,s=5,c='b')
plt.show()
```

```python
def gradient(X,Y,w,b):
    
    dw=2*sum((Y-w*X-b)*(-X))/len(X) 
    db=2*sum((Y-w*X-b)*(-1))/len(X) 
    
    return dw, db

def get_loss(X,Y,w,b):
    return sum((Y-w*X-b)**2)/len(X)

def linear_regression(X,Y,w_init,b_init,alpha,beta):
    dw=1
    db=1
    
    w=w_init
    b=b_init
    
    epsilon=10E-5
    
    W=[]
    B=[]
    COSTS=[]
    
    n_iter=0
    while((abs(dw)>epsilon) and (abs(db)>epsilon)):
        dw, db = gradient(X,Y,w,b)
        
        w=w-alpha*dw
        b=b-beta*db
        
        cost=get_loss(X,Y,w,b)
        
        W.append(w)
        B.append(b)
        COSTS.append(cost)
        n_iter+=1
        
    return W, B, COSTS, n_iter
        
w_init=7
b_init=2
alpha=10E-3
beta=10E-3
W, B, COSTS, n_iter = linear_regression(X,Y,w_init,b_init,alpha,beta)

W[-1], B[-1], COSTS[-1], n_iter
```

```python
fig=plt.figure(figsize=(18,5))

plt.subplot(131)
plt.title('W')
plt.plot(W,c='b')

plt.subplot(132)
plt.title('B')
plt.plot(B,c='b')

plt.subplot(133)
plt.title('COSTS')
plt.plot(COSTS,c='b')
plt.show()
```
