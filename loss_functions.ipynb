```python
import numpy as np
import matplotlib.pyplot as plt
```

### Mean Squared Error Loss

```python
true_value=500
predicted_value=np.arange(0,1000,1)

def mse_loss(true, predicted):
    return (true-predicted)**2

losses = list(map(lambda x: mse_loss(true_value, x), predicted_value))

plt.plot(predicted_value, losses, c='g')
plt.xlabel('predictions')
plt.ylabel('loss')
plt.show()
```

### Root Mean Squared Error Loss

```python
true_value=500
predicted_value=np.arange(0,1000,1)

def rmse_loss(true, predicted):
    return np.sqrt((true-predicted)**2)

losses = list(map(lambda x: rmse_loss(true_value, x), predicted_value))

plt.plot(predicted_value, losses, c='g')
plt.xlabel('predictions')
plt.ylabel('loss')
plt.show()
```

### Mean Absolute Error Loss

```python
true_value=500
predicted_value=np.arange(0,1000,1)

def mae_loss(true, predicted):
    return np.abs((true-predicted))

losses = list(map(lambda x: mae_loss(true_value, x), predicted_value))

plt.plot(predicted_value, losses, c='g')
plt.xlabel('predictions')
plt.ylabel('loss')
plt.show()
```

**RMSE and MAE are same for one example.**

### Exponential Loss

```python
true_value=1
predicted_value=np.arange(-1,10,0.1)

def exponential_loss(true, predicted):
    return np.exp(-1 * true * predicted)

losses = list(map(lambda x: exponential_loss(true_value, x), predicted_value))

plt.plot(predicted_value, losses, c='g')
plt.xlabel('predictions')
plt.ylabel('loss')
plt.show()
```

### Cross Entropy Loss / Negative Log-Likelihood Loss

```python
true_value=1
predicted_value=np.arange(0.1,1,0.01)

def nll_loss(true, predicted):
    return -(true * np.log(predicted) + (1-true) * np.log(1-predicted))

losses = list(map(lambda x: nll_loss(true_value, x), predicted_value))

plt.plot(predicted_value, losses, c='g')
plt.xlabel('predictions')
plt.ylabel('loss')
plt.show()
```

### Logistic Loss

```python
def logistic_loss(true_value, feature_values):
    losses = list(map(lambda x: np.log2(1 + np.exp(x*true_value)), feature_values))
    return losses

true_value = -1
feature_values = np.arange(-5,5,0.1)
losses = logistic_loss(true_value, feature_values)

plt.plot(feature_values, losses, c='g')
plt.show()
```

### 0/1 loss

```python
true_value = 1
margin = np.arange(-5,10,0.1)

def zero_one_loss(true_value, margin):
    if (true_value * margin>0):
        return 0
    else:
        return 1
    
zero_one_loss = list(map(lambda x: zero_one_loss(true_value, x), margin))

plt.plot(margin, zero_one_loss, c='g')
plt.show()
```

### Hinge Loss

```python
true_value = 1
margin = np.arange(-5,10,0.1)

def hinge_loss(true_value, margin):
    return max(0, 1 - true_value * margin)
    
hinge_loss = list(map(lambda x: hinge_loss(true_value, x), margin))

plt.plot(margin, zero_one_loss, c='b')
plt.plot(margin, hinge_loss, c='g')
plt.show()
```
